```python
#importing libraries 
import pandas as pd
import numpy as np
from random import randint
```

```python
# set to display all columns
pd.set_option('display.max_columns', None)
# show all 
```

```python
# import the data 
# first the data from the houses of houses, then houses
houses = pd.read_csv('data/bolha_houses_1-50_15-09-2022_11-36-47.csv')
houses.drop('Unnamed: 0', axis=1, inplace=True)
print('houses: ')
houses.head()


```

```python
# if name of column exists, replace with new name from dictionary
def rename_columns(df, dictionary):
    for key, value in dictionary.items():
        if key in df.columns:
            df.rename(columns={key: value}, inplace=True)
    return df
```

```python
new_column_names = {
  'Cena':           'Price', 
  'Približno':      'ApproximatePrice',
  'Neto cena':      'NetPrice',   
  'DDV':            'Tax', 
  'Bivalna površina':    'SurfaceArea',
  'Vrsta stanovanja':     'ApartmentType', 
  'Leto izgradnje':       'YearBuilt',
  'Vrsta ponudbe':        'OfferType', 
  'Lokacija':             'Location',
  'Umeščenost stanovanja':'ApartmentSetting', 
  'Nadstropje':           'FloorNumber', 
  'Skupno število nadstropij v stavbi':   'TotalFloorsInBuilding', 
  'Leto zadnje prenove':                  'YearRenovated',
  'Balkon / terasa / atrij':    'BalconyTerraceAtrium', 
  'Zunanje površine in oprema':'OutdoorFurniture', 
  'Vrsta cene': 'PriceType',
  'Oglas je objavljen' :'DatePosted',
  'Do poteka še':'ExpireDate',
  'Oglas je prikazan':'AdShowed',  
  'Link': 'Link',
  'Šifra oglasa': 'AdId',  
  'Opremljenost in stanje': 'Furniture', 
  'Število parkirnih mest':'ParkingSpots',
  'Možnost menjave': 'SwitchPosible',
  'Neto površina': 'NetArea',
  'Energetski razred':'EnergyClass',
  'Interna številka oglasa': 'InternalAdID',
  'type': 'ListingType', 
  'Število sob':'TotalRooms',
  'Število nadstropij':'TotalFloors',
  'Vrsta hiše (gradnja)': 'HouseMaterial', 
  'Vrsta hiše': 'HouseType',
  'Površina parcele':'LandArea'
}
```

```python
# use new_column_names dictionary to rename columns
houses = rename_columns(houses, new_column_names)
```

```python
houses.info()
```

```python
# drop columns internaldID and link, DatePosted and ExpireDate, AdShowed, OfferType, , switchPosible, PriceType.
houses.drop(['InternalAdID', 'Link', 'DatePosted', 'ExpireDate', 'AdShowed', 'OfferType', 'SwitchPosible', 'PriceType','AdId'], axis=1, inplace=True)

# drop all features with more than 30% non-null values
houses.dropna(thresh=houses.shape[0]*0.3, axis=1, inplace=True)

# drop listings for which energy class contains string HR
houses = houses[houses['EnergyClass'] != 'HR']
houses.info()
```

```python
# trim the price column and remove . from the price column, set as type integer, replace prices with podogovoru with nan
houses['Price'] = houses['Price'].str.replace('.', '')
houses['Price'] = houses['Price'].str.replace('€', '')
houses['Price'] = houses['Price'].str.replace(' ', '')
houses['Price'] = houses['Price'].str.split(',').str[0]
houses['Price'] = houses['Price'].str.replace('Podogovoru', 'NaN')



# remove all spaces from the column
houses['Price'] = houses['Price'].str.replace(' ', '')

# if 'Nan' in the column, set the value to nan
houses['Price'] = houses['Price'].replace('NaN', np.nan)

# set the column as type integer
houses['Price'] = houses['Price'].fillna(0).astype(np.int64)

# drop all rows with nan values for Price
houses = houses[houses['Price'] != 0]

#  split SurfaceArea by " " and take only the first element, replace , with "", divide by 100 and round to create integer
houses['SurfaceArea'] = houses['SurfaceArea'].str.split(' ').str[0]
houses['SurfaceArea'] = houses['SurfaceArea'].str.replace(',', '')
houses['SurfaceArea'] = houses['SurfaceArea'].astype(float)

houses['SurfaceArea'] = (houses['SurfaceArea'].round(0)/100).round(0).fillna(0).astype(np.int64)

houses['LandArea'] = houses['LandArea'].str.split(' ').str[0]
houses['LandArea'] = houses['LandArea'].str.replace(',', '')
houses['LandArea'] = houses['LandArea'].astype(float)
houses['LandArea'] = (houses['LandArea'].round(0)/100).round(0).fillna(0).astype(np.int64)

#  split NetArea by " " and take only the first element, replace , with "", divide by 100 and round to create integer
houses['NetArea'] = houses['NetArea'].str.split(' ').str[0]
houses['NetArea'] = houses['NetArea'].str.replace(',', '')
houses['NetArea'] = houses['NetArea'].astype(float)
houses['NetArea'] = (houses['NetArea'].round(0)/100).round(0).fillna(0).astype(np.int64)

# convert YearBuilt, YearRenovated to int if not Nan
houses['YearBuilt'] = houses['YearBuilt'].fillna(0).astype(np.int64)
houses['YearRenovated'] = houses['YearRenovated'].fillna(0).astype(np.int64)



# display the first nad last 5 rows of the houses
houses[['Price', 'SurfaceArea', 'NetArea' ,'YearBuilt', 'YearRenovated']].iloc[[*range(5), *range(-5, 0)]]
```

```python
# split location in multiple columns by comma
houses[['Region', 'District', 'Neighborhood']] = houses['Location'].str.split(',', expand=True)
houses.drop('Location', axis=1, inplace=True)
houses.info()
```

```python
# if SurfaceArea is 0, set it to NetArea value
houses.loc[houses['SurfaceArea'] == 0, 'SurfaceArea'] = houses['NetArea']

# remove NetArea column
houses.drop('NetArea', axis=1, inplace=True)
```

```python
# drop all rows with values for Price over 1 million and if price is less than 30000

houses = houses[houses['Price'] > 30000 ]
houses = houses[houses['Price'] < 500000]
```

```python
houses.to_csv('./data/houses_cleaned.csv')
```
