```python
#importing libraries 
import pandas as pd
import numpy as np
from random import randint
```

```python
# set to display all columns
pd.set_option('display.max_columns', None)
# show all 
```

```python
# import the data 
# first the data from the flats of flats, then flats
flats = pd.read_csv('data/bolha_flats_1-43_14-09-2022_23-00-42.csv')
flats.drop('Unnamed: 0', axis=1, inplace=True)
print('flats: ')
flats.head()


```

```python
# if name of column exists, replace with new name from dictionary
def rename_columns(df, dictionary):
    for key, value in dictionary.items():
        if key in df.columns:
            df.rename(columns={key: value}, inplace=True)
    return df
```

```python
new_column_names = {
  'Cena':           'Price', 
  'Približno':      'ApproximatePrice',
  'Neto cena':      'NetPrice',   
  'DDV':            'Tax', 
  'Bivalna površina':    'SurfaceArea',
  'Vrsta stanovanja':     'ApartmentType', 
  'Leto izgradnje':       'YearBuilt',
  'Vrsta ponudbe':        'OfferType', 
  'Lokacija':             'Location',
  'Umeščenost stanovanja':'ApartmentSetting', 
  'Nadstropje':           'FloorNumber', 
  'Skupno število nadstropij v stavbi':   'TotalFloorsInBuilding', 
  'Leto zadnje prenove':                  'YearRenovated',
  'Balkon / terasa / atrij':    'BalconyTerraceAtrium', 
  'Zunanje površine in oprema':'OutdoorFurniture', 
  'Vrsta cene': 'PriceType',
  'Oglas je objavljen' :'DatePosted',
  'Do poteka še':'ExpireDate',
  'Oglas je prikazan':'AdShowed',  
  'Link': 'Link',
  'Šifra oglasa': 'AdId',  
  'Opremljenost in stanje': 'Furniture', 
  'Število parkirnih mest':'ParkingSpots',
  'Možnost menjave': 'SwitchPosible',
  'Neto površina': 'NetArea',
  'Energetski razred':'EnergyClass',
  'Interna številka oglasa': 'InternalAdID',
  'type': 'ListingType', 
  'Število sob':'TotalRooms',
  'Število nadstropij':'TotalFloors',
  'Vrsta hiše (gradnja)': 'HouseMaterial', 
  'Vrsta hiše': 'HouseType',
  'Površina parcele':'LandArea'
}
```

```python
# use new_column_names dictionary to rename columns
flats = rename_columns(flats, new_column_names)
```

```python
flats.info()
```

```python
# drop columns internaldID and link, DatePosted and ExpireDate, AdShowed, OfferType, , switchPosible, PriceType.
flats.drop(['InternalAdID', 'Link', 'DatePosted', 'ExpireDate', 'AdShowed', 'OfferType', 'SwitchPosible', 'PriceType','AdId'], axis=1, inplace=True)

# drop all features with more than 30% non-null values
flats.dropna(thresh=flats.shape[0]*0.3, axis=1, inplace=True)

# drop listings for which energy class contains string HR
flats = flats[flats['EnergyClass'] != 'HR']
flats.info()
```

```python
# trim the price column and remove . from the price column, set as type integer, replace prices with podogovoru with nan
flats['Price'] = flats['Price'].str.replace('.', '')
flats['Price'] = flats['Price'].str.replace('€', '')
flats['Price'] = flats['Price'].str.replace(' ', '')
flats['Price'] = flats['Price'].str.split(',').str[0]
flats['Price'] = flats['Price'].str.replace('Podogovoru', 'NaN')


# remove all spaces from the column
flats['Price'] = flats['Price'].str.replace(' ', '')

# if 'Nan' in the column, set the value to nan
flats['Price'] = flats['Price'].replace('NaN', np.nan)

# set the column as type integer
flats['Price'] = flats['Price'].fillna(0).astype(np.int64)

# drop all rows with nan values for Price
flats = flats[flats['Price'] != 0]

#  split SurfaceArea by " " and take only the first element, replace , with "", divide by 100 and round to create integer
flats['SurfaceArea'] = flats['SurfaceArea'].str.split(' ').str[0]
flats['SurfaceArea'] = flats['SurfaceArea'].str.replace(',', '')
flats['SurfaceArea'] = flats['SurfaceArea'].astype(float)

flats['SurfaceArea'] = (flats['SurfaceArea'].round(0)/100).round(0).fillna(0).astype(np.int64)


#  split NetArea by " " and take only the first element, replace , with "", divide by 100 and round to create integer
flats['NetArea'] = flats['NetArea'].str.split(' ').str[0]
flats['NetArea'] = flats['NetArea'].str.replace(',', '')
flats['NetArea'] = flats['NetArea'].astype(float)
flats['NetArea'] = (flats['NetArea'].round(0)/100).round(0).fillna(0).astype(np.int64)

# convert YearBuilt, YearRenovated to int if not Nan
flats['YearBuilt'] = flats['YearBuilt'].fillna(0).astype(np.int64)
flats['YearRenovated'] = flats['YearRenovated'].fillna(0).astype(np.int64)

# convert TotalFloorsInBuilding to int if not Nan
flats['TotalFloorsInBuilding'] = flats['TotalFloorsInBuilding'].fillna(0).astype(np.int64)

# display the first nad last 5 rows of the flats
flats[['Price', 'SurfaceArea', 'NetArea' ,'YearBuilt', 'YearRenovated']].iloc[[*range(5), *range(-5, 0)]]
```

```python
# split location in multiple columns by comma
flats[['Region', 'District', 'Neighborhood']] = flats['Location'].str.split(',', expand=True)
flats.drop('Location', axis=1, inplace=True)
flats.info()
```

```python
# if SurfaceArea is 0, set it to NetArea value
flats.loc[flats['SurfaceArea'] == 0, 'SurfaceArea'] = flats['NetArea']

# remove NetArea column
flats.drop('NetArea', axis=1, inplace=True)
```

```python
# drop all rows with values for Price over 1 million and if price is below 30 000

flats = flats[flats['Price'] > 30000 ]
flats = flats[flats['Price'] < 500000]
```

```python
flats.to_csv('./data/flats_cleaned.csv')
```
