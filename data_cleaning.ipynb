```python
#importing libraries 
import pandas as pd
import numpy as np
from random import randint
```

```python
# set to display all columns
pd.set_option('display.max_columns', None)
# show all 
```

```python
# import the data 
# first the data from the dataset of flats, then houses
flats = pd.read_csv('data/bolha_flats_1-43_14-09-2022_23-00-42.csv')

# first the data from the dataset of flats, then houses
houses = pd.read_csv('data/bolha_houses_1-50_15-09-2022_11-36-47.csv')

#print('flats: ')
#flats.head()
print('houses: ')
houses.head()

```

```python
# combine the two datasets, drop unnamed index column and show table shape
# add column with value 1 for flats and 0 for houses
flats['type'] = 'flat'
houses['type'] = 'house'
combined_dataset = pd.concat([flats, houses], axis=0, ignore_index=True)
combined_dataset.drop('Unnamed: 0', axis=1, inplace=True)
combined_dataset.shape
```

```python
# write new dataset in the csv format to store it
combined_dataset.to_csv('./data/combined_dataset.csv')
```

```python
#if needed, read the dataset again
dataset = pd.read_csv('./data/combined_dataset.csv')
dataset.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# replace " " in columns names with "_"
dataset.columns = dataset.columns.str.replace(' ', '_')
dataset.info()
```

```python
# show values for the column 'type'
dataset['type'].value_counts()
```

```python
# all column names in a disctionary
# column_names = dataset.columns.to_list()
# we set new, english names for the columns
new_column_names = [
'Price',                #1
'ApproximatePrice',     #2
'NetPrice',             #3   
'Tax',                  #4
'SurfaceArea',          #5
'ApartmentType',       #6
'YearBuilt',          #7
'OfferType',           #8
'Location',           #9    
'ApartmentSetting',  #10
'FloorNumber',          # 11
'TotalFloorsInBuilding', #12
'YearRenovated',        #13
'BalconyTerraceAtrium', #14
'OutdoorFurniture',    #15
'PriceType',          #16
'DatePosted',        #17
'ExpireDate',      #18
'AdShowed',       #19
'Link',           #20
'AdId',         #21
'Furniture',   #22
'ParkingSpots', #23
'SwitchPosible', #24
'NetArea', #25
'EnergyClass', #26
'InternalAdID', #27
'ListingType', #28
'TotalRooms', #28
'TotalFloors', #29
'HouseMaterial', #30
'HouseType', #31
'LandArea'] #32 


# rename the columns to make them more readable
dataset.columns = new_column_names
dataset.info()
```

```python
# drop all features with less than 500 non-null values
dataset.dropna(axis=1, thresh=500, inplace=True)
# drop columns internaldID and link, DatePosted and ExpireDate, AdShowed, OfferType, , switchPosible, PriceType.
dataset.drop(['InternalAdID', 'Link', 'DatePosted', 'ExpireDate', 'AdShowed', 'OfferType', 'SwitchPosible', 'PriceType','AdId'], axis=1, inplace=True)

dataset.info()
```

```python
# check to see how numeric features values look like
dataset[['Price', 'SurfaceArea', 'LandArea', 'YearBuilt', 'YearRenovated']].iloc[[*range(5), *range(-5, 0)]]
```

```python
# drop listings for which energy class contains string HR
dataset = dataset[dataset['EnergyClass'] != 'HR']
```

```python
# trim the price column and remove . from the price column, set as type integer, replace prices with podogovoru with nan
dataset['Price'] = dataset['Price'].str.replace('.', '')
dataset['Price'] = dataset['Price'].str.replace('€', '')
dataset['Price'] = dataset['Price'].str.replace(' ', '')
dataset['Price'] = dataset['Price'].str.split(',').str[0]
dataset['Price'] = dataset['Price'].str.replace('Podogovoru', 'NaN')



# remove all spaces from the column
dataset['Price'] = dataset['Price'].str.replace(' ', '')

# if 'Nan' in the column, set the value to nan
dataset['Price'] = dataset['Price'].replace('NaN', np.nan)

# set the column as type integer
dataset['Price'] = dataset['Price'].fillna(0).astype(np.int64)

# drop all rows with nan values for Price
dataset = dataset[dataset['Price'] != 0]

#  split SurfaceArea and LandArea by " " and take only the first element, replace , with "", divide by 100 and round to create integer
dataset['SurfaceArea'] = dataset['SurfaceArea'].str.split(' ').str[0]
dataset['SurfaceArea'] = dataset['SurfaceArea'].str.replace(',', '')
dataset['SurfaceArea'] = dataset['SurfaceArea'].astype(float)

dataset['SurfaceArea'] = (dataset['SurfaceArea'].round(0)/100).round(0).fillna(0).astype(np.int64)

dataset['LandArea'] = dataset['LandArea'].str.split(' ').str[0]
dataset['LandArea'] = dataset['LandArea'].str.replace(',', '')
dataset['LandArea'] = dataset['LandArea'].astype(float)
dataset['LandArea'] = (dataset['LandArea'].round(0)/100).round(0).fillna(0).astype(np.int64)

# convert YearBuilt, YearRenovated to int if not Nan
dataset['YearBuilt'] = dataset['YearBuilt'].fillna(0).astype(np.int64)
dataset['YearRenovated'] = dataset['YearRenovated'].fillna(0).astype(np.int64)

# convert TotalFloorsInBuilding to int if not Nan
dataset['TotalFloorsInBuilding'] = dataset['TotalFloorsInBuilding'].fillna(0).astype(np.int64)

# display the first nad last 5 rows of the dataset
dataset[['Price', 'SurfaceArea', 'LandArea', 'YearBuilt', 'YearRenovated']].iloc[[*range(5), *range(-5, 0)]]
```

```python
# split location in multiple columns by comma
dataset[['Region', 'District', 'Neighborhood']] = dataset['Location'].str.split(',', expand=True)
dataset.drop('Location', axis=1, inplace=True)
dataset.info()
```

```python
# take + out of the column parking spots and replace string brez pripadajočega parkirišča with 0
dataset['ParkingSpots'] = dataset['ParkingSpots'].str.replace('+', '')
dataset['ParkingSpots'] = dataset['ParkingSpots'].str.replace('brez pripadajočega parkirišča', '0')
dataset['ParkingSpots'] = dataset['ParkingSpots'].fillna(0).astype(np.int64)

# take + out of the column TotalRooms and set type to int
dataset['TotalRooms'] = dataset['TotalRooms'].str.replace('+', '')
dataset['TotalRooms'] = dataset['TotalRooms'].fillna(0).astype(np.int64)

# set apartment type string -sobno to -bedroom
dataset['ApartmentType'] = dataset['ApartmentType'].str.replace('sobno', 'bedroom')
dataset['ApartmentType'] = dataset['ApartmentType'].str.replace('Garsonjera', 'Studio')


# drop all rows with values either Mansarda, Pritličje, visoko pritličje or klet
dataset = dataset[dataset['FloorNumber'] != 'Mansarda']
dataset = dataset[dataset['FloorNumber'] != 'Pritličje']
dataset = dataset[dataset['FloorNumber'] != 'Visoko pritličje']
dataset = dataset[dataset['FloorNumber'] != 'Klet']

dataset = dataset.fillna('missing')
```

```python
# drop all rows with values for Price over 1 million
dataset = dataset[dataset['Price'] < 1000000]
```
